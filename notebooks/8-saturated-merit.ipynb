```python
import math
import itertools
import pandas
```

```python
pandas.set_option('display.max_rows', None)
```

```python
enum = { "nada": 0, "básico": 1, "médio": 2, "avançado": 3 }
inverse_enum = { 0: "nada", 1: "básico", 2: "médio", 3: "avançado" }
r = itertools.combinations_with_replacement([
    enum["nada"],
    enum["básico"],
    enum["médio"],
    enum["avançado"]
], 3)
r = list(r)
```

```python
x = [enum["nada"],
    enum["básico"],
    enum["médio"],
    enum["avançado"]]


r = [p for p in itertools.product(x, repeat=3)]

r = list(r)
```

```python
def format_result(i, sim, idx = ""):
    return {
        f"Excel{idx}": inverse_enum[i[0]],
        f"PPT{idx}": inverse_enum[i[1]],
        f"Word{idx}": inverse_enum[i[2]],
        f"Similaridade{idx}": sim
    }
```

```python
def exponential_sim_match(i, j = [3, 3, 3], C = [1, 1, 1], idx = ""):
    numerator = 0
    denominator = 0
    
    for _i, _j, _c in list(zip(i, j, C)):        
        _numerator = ((_i/3)**2 - (_j/3)**2)*((_j/3)+1)
        
        if _i > _j:
            numerator += _numerator*_c
        else:
            numerator += _numerator
        
        denominator += ((_j/3) + 1)        
        
    return format_result(i, '%f' % (numerator/denominator), idx)
```

```python
exponential_sim_match([3, 3, 1], [0, 0, 0], [0.1, 0.1, 0.1]), exponential_sim_match([3, 1, 1], [0, 0, 0], [0.1, 0.1, 0.1]), exponential_sim_match([1, 1, 1], [0, 0, 0], [0.1, 0.1, 0.1])
```

```python
def binary_merit_sim_match(i, j = [3, 3, 3], C = [1, 1, 1], idx = ""):
    numerator = 0
    denominator = 0
    
    consider_merit = False
    ct_zeroes = C.count(0)
    if ct_zeroes > (len(i) // 2): # // rounds down e.g. 3 // 2 = 1
        consider_merit = True
        
    for _i, _j, _c in list(zip(i, j, C)):
        _numerator = ((_i/3)**2 - (_j/3)**2)*((_j/3)+1)
        
        if _i > _j:
            if consider_merit:
                numerator += _numerator
            else:
                numerator += _numerator*_c
        else:
            numerator += _numerator
        
        denominator += ((_j/3) + 1)        
        
    return format_result(i, '%f' % (numerator/denominator), idx)
```

```python
[binary_merit_sim_match([2, 2, 2, 0], [2, 2, 0, 2], [1, 1, 0, 0]),
 binary_merit_sim_match([2, 2, 2, 1], [2, 2, 0, 2], [1, 1, 0, 0]),
 binary_merit_sim_match([2, 2, 2, 2], [2, 2, 0, 2], [1, 1, 0, 0]),
 binary_merit_sim_match([2, 2, 2, 3], [2, 2, 0, 2], [1, 1, 0, 0])]
```

```python
j_s1 = [2, 2, 2]
c_s1 = [1, 1, 0]
result = list(map(lambda i: exponential_sim_match(i, j_s1, c_s1, 1), list(r)))

j_s2 = [2, 2, 2]
c_s2 = [1, 1, 1]
result2 = list(map(lambda i: exponential_sim_match(i, j_s1, c_s2, 2), list(r)))
```

```python
result = sorted(result, key=lambda x: x["Similaridade1"])
result2 = sorted(result2, key=lambda x: x["Similaridade2"])
```

```python
for i, v in enumerate(result): result[i]['idx1'] = i
for i, v in enumerate(result): result2[i]['idx2'] = i
```

```python
result1_df = pandas.DataFrame(result)
result2_df = pandas.DataFrame(result2)

print(f"Empresa1: {str(j_s1)} | c1 = {str(c_s1)} ======= Empresa2: {str(j_s2)} | c2 = {str(c_s2)} ")
pandas.concat([result1_df, result2_df], axis=1)
```
